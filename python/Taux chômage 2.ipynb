```python
# Importation de la librairie pandas avec alias
import pandas as pd
```

```python
#Lecture du fichier avec sélection de la 2ème feuille et saut des 5 premières lignes
data = pd.read_excel('chomage_zone.xls', sheet_name=1, header=0, skiprows=[0,1,2,3,4])
data
```

```python
#Taux de chômage moyen par trimestre et par région
data_grp = data.groupby(['LIBREG'], as_index=False).mean()
data_grp
```

```python
#Diagramme à barres pour le taux de chômage du premier trimestre 2020 par région
data_grp.plot('LIBREG', '2020-T1', 'bar', title='Taux de chômage par région', grid=True, figsize=(20,10))
```

```python
#Isolation des colonnes 3 à 73
data2 = data.iloc[:, 3:74]
data2
```

```python
data_grp2 = data2.groupby(['LIBREG']).mean()
data_grp2
```

```python
#Transposée
data_grp2 = data_grp2.transpose()
data_grp2
```

```python
#Réindexation
data_grp2.reset_index(inplace=True)
data_grp2
```

```python
#Renommage de la colonne index en péridoe
data_grp2.rename({"index": "periode"}, axis='columns', inplace=True)
data_grp2
```

```python
#Evolution du taux de chômage pour la région PACA
data_grp2.plot("periode", "Provence-Alpes-Côte d'Azur", 'line', title='Taux de chômage en région PACA', grid=True, figsize=(20,10))
```

```python
#Evolution du taux de chômage pour chaque région
data_grp2.plot("periode", None, 'line', title='Taux de chômage par région', grid=True, figsize=(20,10), rot=90)
```

```python
#Evolution du taux de chômage pour les régions PACA, Occitanie et IdF
data_grp2.plot("periode", ["Provence-Alpes-Côte d'Azur", "Occitanie", "Ile-de-France"], 'line', title='Taux de chômage pour les régions PACA, Occitanie et Île-de-France', grid=True, figsize=(20,10), rot=90)
```
