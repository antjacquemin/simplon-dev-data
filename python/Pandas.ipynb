```python
#Importation d'un jeu de données Excel 

nom_colonnes = ['consommation', 'cylindres', 'deplacement', 'puissance', 'poids', 'acceleration', 'annee', 'origine', 'nom']
# data = pd.read_excel('//home//simplon//python//autompg.xlsx', index_col=None, header=None, names=nom_colonnes)
data = pd.read_excel('autompg.xlsx', index_col=None, header=None, names=nom_colonnes, skiprows=1)
```

```python
#Pour faire apparaître les graphiques dans le carnet
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Histogramme pour la puissance
data.hist(column='puissance')
```

```python
#Nuage de points pour la consommation en fonction de la puissance
#avec niveaux de gris pour l'année
data.plot.scatter(x='puissance', y='consommation', c='annee')
```

```python
pd.plotting.scatter_matrix(data.select_dtypes(exclude=['object']), figsize=(10,10))
```

```python
#Consommation moyenne des voitures 8 cylindres
print(data.loc[data['cylindres']==8].mean()['consommation'])
#Noms des véhicules des années 70 avec la puissance max
puissanceMax = data.loc[data['annee']==70,'puissance'].max()
data70 = data.loc[data['annee']==70]
print(data70.loc[data70['puissance']==puissanceMax, 'nom'])
```

```python
#Nuage de points pour la puissance en fonction du poids des voitures 6 cylindres
data6c = data.loc[data['cylindres']==6]
data6c.plot.scatter(x='puissance', y='poids')
```
